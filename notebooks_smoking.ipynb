```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy.linalg as la
from sklearn.decomposition import PCA
```

## Smoking and Cancer analysis

This data analysis aims to determine correlation between smoking and cancer. The data can be sourced at https://www.statcrunch.com/reports/view?reportid=36865&tab=preview. It is imported into this notebook as data_project.csv. To analyze this dataset, I will be applying a least squares regression for 4 different types of cancer: lung cancer, bladder cancer, kidney cancer, and leukhemia. The reason for choosing these four relationships is because cancer-related deaths can be seen as proportional to the number of smokers. By determining a least squares regression and pearson correlation coefficient for all four plots, the accuracy of this statement can be determined.

```python
data = pd.read_csv('data_project.csv')

cig = data['CIG']
lung = data['LUNG']
blad = data['BLAD']
kid = data['KID']
leuk = data['LEUK']

# Create a scatter plot for CIG vs LUNG
plt.scatter(cig, lung)
plt.xlabel('Cigarettes Sold per Capita')
plt.ylabel('Lung Cancer per Capita')
plt.title('Cigarettes Sold vs Lung Cancer')
plt.show()

# Create a scatter plot for CIG vs BLAD
plt.scatter(cig, blad)
plt.xlabel('Cigarettes Sold per Capita')
plt.ylabel('Bladder Cancer per Capita')
plt.title('Cigarettes Sold vs Bladder Cancer')
plt.show()

# Create a scatter plot for CIG vs KID
plt.scatter(cig, kid)
plt.xlabel('Cigarettes Sold per Capita')
plt.ylabel('Kidney Cancer per Capita')
plt.title('Cigarettes Sold vs Kidney Cancer')
plt.show()

# Create a scatter plot for CIG vs BLAD
plt.scatter(cig, leuk)
plt.xlabel('Cigarettes Sold per Capita')
plt.ylabel('Leukhemia per Capita')
plt.title('Cigarettes Sold vs Leukhemia')
plt.show()
```

## Regression on Lung Cancer Data

```python
cigs_per_capita = np.array(cig)
lung_per_capita = np.array(lung)

regression = np.column_stack((cigs_per_capita, np.ones_like(cigs_per_capita)))
coefficients, residuals, _, _ = la.lstsq(regression, lung_per_capita, rcond = None)

slope, intercept = coefficients

# Calculate Pearson Correlation Coefficient
correlation = np.corrcoef(cig, lung)[0, 1]

# Calculate R Squared

y_pred = intercept + slope * cig

residuals = lung_per_capita - y_pred
ssr = np.sum(residuals ** 2)
y_mean = np.mean(lung_per_capita)
sst = np.sum((lung_per_capita - y_mean) ** 2)
r_squared = 1 - (ssr / sst)

plt.scatter(cig, lung, label='Data')

# Plot the regression line
plt.plot(cig, slope * cig + intercept, color='red', label='Regression Line')
plt.annotate(f'Correlation: {correlation:.3f}', xy=(0.5, 0.9), xycoords='axes fraction', fontsize=10)
plt.annotate(f'R-squared: {r_squared:.3f}', xy=(0.5, 0.8), xycoords='axes fraction', fontsize=10)
plt.xlabel('Cigarettes Sold per Capita')
plt.ylabel('Lung Cancer per Capita')
plt.title('Linear Regression - Cigarettes Sold vs Lung Cancer')
plt.legend()
plt.show()
```

```python
cancers = ['LUNG', 'BLAD', 'KID', 'LEUK', 'CIG']  # Add more cancer types if necessary

subset = data[cancers]

pca = PCA(n_components=len(cancers))
pca.fit(subset)

loadings = pd.DataFrame(pca.components_, columns=cancers, index=['PC'+str(i+1) for i in range(len(cancers))])

# Plot the loadings
loadings.plot(kind='bar', rot=0)
plt.xlabel('Principal Components')
plt.ylabel('Loadings')
plt.title('Principal Component Loadings')
plt.show()
```

## Analysis
<p>
In plotting all four cancers and performing a regression on all of them, it becomes apparent that the most correlated to smoking are lung cancer and bladder cancer. They both have the highest correlation coefficients with the difference from the next best being 0.110, and also the highest R-squared values. Of the two, Bladder seems to be more correlated by the correlation coefficient and r squared value, but it is difficult to discern this minimal of a difference with such variable data. This is further supported by the principal component analysis, which shows that lung cancer and cigarette usage are a significant component. Although this method doesn't exactly explain the reasons for the principal component, this could be intepreted as linearity, which matches up well with the data. Other interesting insights from the PCA are that Leukhemia often has very negative loadings, which support the idea that its data is relatively random and variable, lending to no particular pattern.
