```python
import pandas as pd
```

```python
data = pd.read_csv('data_project.csv')

columns_to_display = data.columns[2:]

for column in columns_to_display:
    styled_table = data[[data.columns[0], data.columns[1], column]].style

    styled_table.background_gradient(cmap='Blues')
    styled_table.set_properties(**{'text-align': 'center'})

    # Display the styled DataFrame
    display(styled_table)
```
